```python
print("Structure:")
df.info()
print("\n" + "="*50 + "\n")

print("Countries:")
print(df['Country'].value_counts().head(10))
print("\n" + "="*50 + "\n")

print("Age groups:")
print(df['Age'].unique())
print("\n" + "="*50 + "\n")

print(f"Duplicates: {df.duplicated().sum()}")
print("\n" + "="*50 + "\n")

print(f"Missing values: \n {df.isnull().sum()}")
print("\n" + "="*50 + "\n")
```

### Завдання 1. Підрахунок загальної кількості респондентів

```python
total_res = df.shape[0]
total_res
```

### Завдання 2. Аналіз повноти відповідей респондентів

```python
schema = set(schema_raw["qname"])
df_col = set(df.columns)
cols = list(schema & df_col)
df.dropna(subset=cols).shape[0]
```

Жоден респонден невідповів на всі питання

### Завдання 3. Статистичний аналіз досвіду респондентів

```python
work_mean = df['WorkExp'].mean()
work_median = df['WorkExp'].median()
work_mode = df['WorkExp'].mode()[0]

print(f"Mean: {work_mean:.2f}")
print(f"Median: {work_median}")
print(f"Moda: {work_mode}")
```

### Завдання 4. Аналіз віддаленої роботи

```python
rem = df['RemoteWork'] == 'Remote'
df[rem].shape[0]
```

### Завдання 5. Визначення популярності Python

```python
p_users = df['LanguageHaveWorkedWith'].str.contains('Python', na=False)
p_users_count = df[p_users].shape[0]
p_users_pr = (p_users_count/total_res) * 100
print(f"Python users: {p_users_pr:.2f}%")
```

### Завдання 7. Географічний аналіз компенсації Python-розробників

```python
python_df = df[df['LanguageHaveWorkedWith'].str.contains('Python', na=False)].copy()
```

```python
salary_by_country = python_df.groupby('Country')['ConvertedCompYearly'].agg(['mean', 'median'])
salary_by_country['count'] = python_df.groupby('Country')['ConvertedCompYearly'].count()
salary_by_country = salary_by_country.sort_values(by='median', ascending=False)

top_countries = salary_by_country[salary_by_country['count'] > 50]
top_countries.style.format("{:.2f}")
```
